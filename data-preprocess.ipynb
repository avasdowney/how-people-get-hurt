```python
# imports needed
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
# import NEISS.xlsx
data = pd.read_excel("datasets/NEISS.xlsx")
# data = pd.read_csv('datasets/small.csv')

# Read the DataFrame, first using the feature data
features = ["CPSC_Case_Number", "Treatment_Date", "Age", "Sex",	"Race",	"Body_Part", "Diagnosis",	
    "Disposition", "Location", "Fire_Involvement", "Product_1", "Product_2", "Product_3", 
    "PSU",	"Weight"]
df = pd.DataFrame(data, columns=features)
del data
df.head(5)
```

```python
scaler = MinMaxScaler()
data = pd.DataFrame(scaler.fit_transform(df))
data
```

```python
# split into train and test sets
train, test = train_test_split(df, test_size=0.33)
del df
```

```python
# save train and test sets to excel files
train.to_csv("datasets/train.csv")
test.to_csv("datasets/test.csv")
```
